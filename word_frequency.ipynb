```python
import jieba
import codecs
import jieba.posseg as pseg
import pandas as pd

#定义一个空词典，用于储存“词-词频”的键值对
dict1={}
f=codecs.open('ip_merged.txt','r',encoding='utf8')

#循环读取文件中的每一行文本
for line in f.readlines():
    #清洗文本
    line=line.strip('/ /\n').split('/')

    #遍历拆分过后的每次词语，统计词频，如果不在字典中，新增键值对，初始词频为1
    for item in line:
        if item not in dict1.keys():
            dict1[item]=1
        #如果已存在，则词频+1
        else:
            dict1[item]+=1
        
    
#对字典按词频降序排序
dict2=sorted(dict1.items(), key=lambda d:d[1],reverse=True)



#定义空列表，用于储存筛选后的结果
result=[]

#遍历排序后前50的词语
for item,item2 in dict2[:50]:
    if len(item)>1:
        result.append((item,item2))

#将result列表转化为DataFrame的格式，并写入excel表格
df=pd.DataFrame(result,columns=['省份','词频'])
df.to_excel('ip_frequency.xlsx',sheet_name='词频结果',index=False)
print("已成功写入")
```
