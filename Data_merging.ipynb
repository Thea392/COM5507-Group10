```python
file_paths = [
    "/Users/sylvie/Documents/5507类型/奇幻玄幻/merge_奇幻_cleaned.csv",
    "/Users/sylvie/Documents/5507类型/古代言情/merge_gy_cleaned.csv",
    "/Users/sylvie/Documents/5507类型/现代言情/merge_xy_cleaned.csv",
    "/Users/sylvie/Documents/5507类型/家庭伦理/merge_jt_cleaned.csv",
    "/Users/sylvie/Documents/5507类型/逆袭/merge_nx_cleaned.csv",
    "/Users/sylvie/Documents/5507类型/萌宝/merge_mb_cleaned.csv"
]
output_file = r"/Users/sylvie/Documents/5507类型/merge_total.csv" 
```

```python
all_data = []
for file_path in file_paths:
     try:
        df = pd.read_csv(file_path)
        df = df.dropna(how='all')  # 删除完全空白的行
        all_data.append(df)
     except Exception as e:
         pass
```

```python
if all_data:
    merged_df = pd.concat(all_data, ignore_index=True, sort=False)
    merged_df_cleaned = merged_df.drop_duplicates(keep='first')  # 去重
    merged_df_cleaned.to_csv(output_file, index=False)
    print(f"{len(merged_df_cleaned)}，保存至：{output_file}")
else:
    print("处理失败")
```
