```python
import codecs
from gensim import corpora
from gensim.models import LdaModel
from gensim.corpora import Dictionary
# import pyLDAvis
# import pyLDAvis.gensim_models as gensimvis
# from pyLDAvis import gensim
from gensim import corpora, models
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis

train = []

fp = open(r"保姆现形记_output.txt",'r',encoding='utf8').readlines()
# for line in fp:
for line in fp:
    line1 = line.strip()
    words = line1.split('/')
    train.append([w for w in words if len(w)>=2])
    # if line != '':
    #     line = line.split()
    #     train.append([w for w in line])

dictionary = corpora.Dictionary(train)

corpus = [dictionary.doc2bow(text) for text in train]

lda = LdaModel(corpus=corpus, id2word=dictionary, num_topics=5, passes=100)

for topic in lda.print_topics(num_words = 80):
    termNumber = topic[0]
    print(topic[0], ':', sep='')
    listOfTerms = topic[1].split('+')
    for term in listOfTerms:
        listItems = term.split('*')
        print('  ', listItems[1], '(', listItems[0], ')', sep='')
```

```python
import pyLDAvis.gensim_models
d=gensimvis.prepare(lda,corpus,dictionary)
pyLDAvis.displace(d)
pyLDAvis.show(d,local=False)
d=gensimvis.prepare(lda,corpus,dictionary)
pyLDAvis.save_html(d,'LDA_pass.html')
```
