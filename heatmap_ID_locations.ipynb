```python
import pandas as pd
import numpy as np
from pyecharts import options as opts
from pyecharts.charts import Map
```

```python
data = pd.read_excel('/Users/sylvie/Documents/5507类型/热力图/ip_frequency.xlsx')
data
```

```python
c = (
    Map(init_opts=opts.InitOpts(width="1400px", height="700px", theme="purple-passion",bg_color="#D0D0D0"))
    .add(
        series_name="",
        data_pair=[list(z) for z in zip(data['provincial-level administrative region'], data['frequency'])],
        maptype="china",
        is_map_symbol_show=False,
        label_opts=opts.LabelOpts(is_show=False)  

    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title="Geographical Distribution of Commenters' IP Locations",
            title_textstyle_opts=opts.TextStyleOpts(
                font_size=20,
                color="#000000"
            ),
            pos_left="center",
            pos_top="3%"
        ),
        legend_opts=opts.LegendOpts(is_show=False),
        visualmap_opts=opts.VisualMapOpts(
            max_=3500,
            is_piecewise=True,
            pieces=[
                {"min": 3000, "label": ">3000"},
                {"min": 2000, "max": 2999, "label": "2000-2999"},
                {"min": 1500, "max": 1999, "label": "1500-1999"},
                {"min": 1000, "max": 1499, "label": "1000-1499"},
                {"min": 500, "max": 999, "label": "500-999"},
                {"min": 200, "max": 499, "label": "200-499"},
                {"min": 1, "max": 199, "label": "1-199"},
            ],
            range_text=["Frequency", ""],
            range_color=["#F0F8FF", "#E0F0FF", "#C0D8FF", "#A0C0FF", "#6090FF", "#3070FF", "#0048FF"],
            pos_left="25%",
            pos_top="45%",
            textstyle_opts=opts.TextStyleOpts(color="#333333", font_size=9), 
            item_width=12,  
            item_height=12, 
        )
    )
)
c.render("./用户分布.html")
```
